```python
import sys
import os
sys.path.append(os.path.abspath("../"))

from scripts.crema_dataset import CREMADataset
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import librosa.display  #  required to visualize spectrograms

dataset = CREMADataset("../data/crema_labels.csv")
loader = DataLoader(dataset, batch_size=4, shuffle=True)

# Test batch
spectrograms, labels = next(iter(loader))
print("Shape of spectrograms:", spectrograms.shape)  # [4, 1, 64, ~time]
print("Labels:", labels)

# Plot Mel Spectrogram of the first sample
plt.figure(figsize=(10, 4))
librosa.display.specshow(spectrograms[0][0].numpy(), sr=16000, x_axis='time', y_axis='mel')
plt.title("Mel Spectrogram (dB) - Sample 1")
plt.colorbar(format="%+2.0f dB")
plt.tight_layout()
plt.show()


```

```python
import sys
import os
sys.path.append(os.path.abspath("../"))

import torch
from scripts.cnn_model import EmotionCNN

model = EmotionCNN(num_classes=6)
output = model(spectrograms)
print("Output shape:", output.shape)  # Should be [4, 6]
```
